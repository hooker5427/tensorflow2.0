```python
path ="./data/titanic/train.csv"
data = pd.read_csv( path ) 
data.head()
```

```python
data.dtypes 
```

```python
data, y =  data , data.pop("survived")
```

```python
category_data  = data.loc[:, data.dtypes == 'object']
numeric_data =  data.loc[:, data.dtypes != 'object']
```

```python
for   col in category_data.columns :
    category_data[col] = pd.Categorical( category_data[col]).codes 
```

```python
category_data.describe()
```

```python
numeric_data.describe()
```

```python
data =pd.concat(   [data.drop(data.columns , axis = 1 ) ,  category_data , numeric_data ] , axis  =1 )
```

```python
data
```

```python
one_hot_cols  =  [ 'sex' ,'embark_town']
one_hot_data = data.loc[:, one_hot_cols  ]
one_hot_data_encoded = pd.get_dummies( one_hot_data , columns=one_hot_cols  )
```

```python
one_hot_data_encoded
```

```python
import seaborn as sns 

plt.figure(figsize=( 12 ,12  ))
plt.title('Pearson Correlation between Features',y=1.05,size=15)
train_data_hot_encoded = data.drop('embark_town',1).drop('sex',1).join(one_hot_data_encoded)
# 计算特征之间的Pearson系数，即相似度
sns.heatmap(train_data_hot_encoded.astype(float).corr(),
            linewidths=0.1,vmax=1.0,
            square=True,
            linecolor='white',
            annot=True )
plt.show()
```
